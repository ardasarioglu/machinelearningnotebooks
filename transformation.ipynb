```python
import math
def plot_all_histograms(df):
    num_cols = df.select_dtypes(include=np.number).columns
    n_cols = 3
    n_rows = math.ceil(len(num_cols) / n_cols)

    plt.figure(figsize=(5*n_cols, 4*n_rows))

    for i, col in enumerate(num_cols, 1):
        plt.subplot(n_rows, n_cols, i)
        sns.histplot(df[col], kde=True, bins=30)
    plt.tight_layout()
    plt.show()
```

```python
plot_all_histograms(df)
```

```python
X = df.drop("MEDV", axis=1)
y = df["MEDV"]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15)
```

```python
from sklearn.preprocessing import PowerTransformer
pt_X = PowerTransformer(method="yeo-johnson")
X_train_transformed = pt_X.fit_transform(X_train)
X_test_transformed = pt_X.transform(X_test)
```

```python
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']

X_train_transformed = pd.DataFrame(X_train_transformed, columns=column_names)
```

```python
from scipy.stats import boxcox
```

```python
y_train.describe()
```

```python
y_train_transformed, y_train_lambda = boxcox(y_train)
```

```python
def inverse_boxcox(y, lambda_):
    if lambda_ == 0:
        return np.exp(y)
    else:
        return np.power(y * lambda_ + 1, 1 / lambda_)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train_transformed, y_train_transformed)
y_pred_transformed = model.predict(X_test_transformed)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
y_pred_original = inverse_boxcox(y_pred_transformed, y_train_lambda)
```

```python
print(f"r2: {r2_score(y_test, y_pred_original)}")
print(f"mse: {mean_squared_error(y_test, y_pred_original)}")
```
