```python
df = df.drop(["deck","embark_town","alive"], axis=1)
```

```python
df["age"] = df["age"].fillna(df["age"].median())
```

```python
df["embarked"] = df["embarked"].fillna(df["embarked"].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df["adult_male"] = df["adult_male"].astype(int)
df["alone"] = df["alone"].astype(int)
```

```python
df.info()
```

```python
X = df.drop("survived", axis=1)
y = df["survived"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=15)
```

```python
from sklearn.compose import ColumnTransformer
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
X.info()
```

```python
for col in X.columns:
    print(col)
    print((X[col].unique()))
    print("------------------------------------")
```

```python
categorical_cols = ["sex","embarked","who","class"]
```

```python
preprocessor = ColumnTransformer(transformers=[
    (
        "cat",OneHotEncoder(drop="first", handle_unknown="ignore"), categorical_cols
    )
], remainder="passthrough"
)
```

```python
X_train = preprocessor.fit_transform(X_train)
X_test = preprocessor.transform(X_test)

X_train = pd.DataFrame(X_train, columns=preprocessor.get_feature_names_out())
X_test = pd.DataFrame(X_test, columns=preprocessor.get_feature_names_out())
```

```python
import lightgbm as lgb
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
clf = lgb.LGBMClassifier(verbosity = -1)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
feature_importance = pd.DataFrame(
    {
        "Feature": X_train.columns,
        "Importance": clf.feature_importances_
    }
)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
lgb_model = lgb.LGBMClassifier(verbosity = -1)
```

```python
params = {
    "n_estimators": [100, 300, 500, 100],
    "max_depth": [3, 5, 7, -1],
    "learning_rate": [0.01, 0.05, 0.1 ,0.3],
    "num_leaves": [15, 31, 63, 127],
    "min_child_samples": [10,20,30],
    "subsample": [0.6, 0.8, 1],
    "colsample_bytree": [0.6,0.8,1]
}
```

```python
cv = RandomizedSearchCV(estimator=lgb_model, param_distributions=params)
```

```python
cv.fit(X_train, y_train)
y_pred = cv.predict(X_test)
```

```python
cv.best_params_
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```
