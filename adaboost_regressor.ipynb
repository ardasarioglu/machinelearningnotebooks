```python
X = df.drop("selling_price", axis=1)
y = df["selling_price"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=15)
```

```python
cat_cols = df.select_dtypes("object").columns.to_list()
```

```python
cat_cols
```

```python
X
```

```python
df[cat_cols].nunique()
```

```python
onehot_columns = ["seller_type", "fuel_type","transmission_type"]
```

```python
freq_columns = ["car_name","brand","model"]
```

```python
for col in freq_columns:
    freq = X_train[col].value_counts() / len(X_train)
    X_train[col + "_freq"] = X_train[col].map(freq)
    X_test[col + "_freq"] = X_test[col].map(freq)

    mean_freq = freq.mean()
    X_test[col+"_freq"] = X_test[col+"_freq"].fillna(mean_freq)
```

```python
X_train
```

```python
X_train.drop(["car_name","brand","model"], axis=1, inplace=True)
X_test.drop(["car_name","brand","model"], axis=1, inplace=True)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(
    transformers=[
        ("onehot",OneHotEncoder(drop="first", handle_unknown="ignore"), onehot_columns)
    ], remainder="passthrough"
)
```

```python
X_train = transformer.fit_transform(X_train)
X_test = transformer.transform(X_test)
```

```python
encoded_cols = transformer.get_feature_names_out()
```

```python
X_train
```

```python
X_train = pd.DataFrame(X_train, columns=encoded_cols)
X_test = pd.DataFrame(X_test, columns=encoded_cols)
```

```python
X_train
```

```python
X_train.info()
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import RandomizedSearchCV
```

```python
model = AdaBoostRegressor()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(mean_absolute_error(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
print(r2_score(y_test, y_pred))
```

```python
params = {
    "n_estimators": [50,80,100,120],
    "learning_rate": [0.001, 0.01, 0.1,1 ,2],
    "loss": ["linear","square","exponential"]
}
```

```python
random = RandomizedSearchCV(estimator=AdaBoostRegressor(), param_distributions=params, scoring="r2", cv=5, )
```

```python
random.fit(X_train, y_train)
y_pred = random.predict(X_test)
```

```python
print(mean_absolute_error(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
print(r2_score(y_test, y_pred))
```
